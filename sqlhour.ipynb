```python
%%sql
-- 1. Find albums that have more than 10 tracks
SELECT 
    a.Title as Album_title,
    count(t.TrackId) as Total_count
FROM 
    albums as a
JOIN
    tracks as t
ON 
    a.AlbumId = t.AlbumId
GROUP BY
    t.AlbumId
HAVING
    count(t.TrackId) > 10
LIMIT 5
```

```python
%%sql
-- 2. List the top 5 customers who spent the most on Rock music
SELECT 
    c.FirstName || " " || c.LastName as Customer,
    g.Name,
    (item.UnitPrice * item.Quantity) as Rock_Spending
FROM 
    customers as c
JOIN 
    invoices as i
ON
    c.CustomerId = i.CustomerId
JOIN
    invoice_items as item
ON 
    i.InvoiceId = item.InvoiceId
JOIN
    tracks as t
ON
    t.TrackId = item.TrackId  
JOIN 
    genres as g
ON 
    g.GenreId = t.GenreId
WHERE 
    g.Name = "Rock"
ORDER BY 
    Rock_Spending DESC
LIMIT 4
```

```python
%%sql
-- 3. Find tracks that are longer than the average track length
SELECT 
    Name as Tracks_name,
    Milliseconds as Duration
FROM 
    tracks
WHERE Milliseconds > (
    SELECT avg(Milliseconds)
    FROM tracks
)
LIMIT 5
```

```python
%%sql
--  4. List all customers and the total number of invoices they have
SELECT 
    c.FirstName || ' ' || c.LastName AS Customer_Name,
    COUNT(i.InvoiceId) AS Invoice_Count
FROM 
    customers c
LEFT JOIN 
    invoices i 
ON 
    c.CustomerId = i.CustomerId
GROUP BY 
    c.CustomerId
ORDER BY 
    Invoice_Count DESC
LIMIT 5
```

```python
%%sql
-- 5. Get the most popular media type (by number of tracks sold)
SELECT 
    m.Name as Media_type,
    count(item.TrackId) As Tracks_sold
FROM 
    media_types as m
JOIN 
    tracks as t
ON
    M.MediaTypeId = T.MediaTypeId
JOIN 
    invoice_items as item
ON
    t.TrackId = item.TrackId
GROUP BY
    m.Name
ORDER BY
    Tracks_sold DESC  
LIMIT 5
```
